```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# set dataframe equal to the 'sets' Excel file
df = pd.read_csv("sets.csv")
```

```python
df.head()
```

```python
# filter rows using .loc
df.loc[df['Theme'] == 'Samsonite']
```

```python
# overview of integer values in dataframe
df.describe()
```

```python
# move Theme column to be placed before num_parts column
cols = list(df.columns)
dftest = df[cols[0:4] + [cols[-1]] + cols[4:5]]
dftest.head()
```

```python
# transfer results to main dataframe
df = dftest
```

```python
df.loc[(df['year'] == 1965)]
```

```python
df.loc[(df['year'] == 2023) & (df['num_parts'] > 500)]
```

```python
# find the sets with the most pieces
top10 = df.sort_values('num_parts', ascending = False)
top10.head(10)
```

```python
# find names of sets that contain 'ci' at any point of the string
df.loc[(df['name'].str.contains('ci[a-z]*', case = False, regex = True)) & (df['num_parts'] > 100)]
```

```python
# determine whether a set is big or not
bigset = df['num_parts'] > 500
df2 = df.assign(big_set = bigset)
df2.head()
```

```python
df2 = df
```

```python
# make a new column that will aid in future groupings
df['count'] = 1
df.groupby('year').count()['count'].sort_values(ascending = False)
```

```python
#find mean of number values in df, but filter to only show mean values for 'num_parts' by year
groups = df.groupby(['year']).mean(numeric_only = True)
avg_num_parts = groups['num_parts']
avg_num_parts
```

```python
avg_num_parts.dtypes
```

```python
#plot values from avg_num_parts to make a line graph
plt.plot(avg_num_parts, color = 'r', linewidth = 1, marker = ".")
plt.title('Average Lego Pieces in Each Set per Year', fontsize = 15)

plt.grid(axis = 'both', linestyle = '-')
plt.xlabel('Year', fontsize = 12)
plt.ylabel('Avg Pieces per Set', fontsize = 12)

plt.show()
```

```python
#find the most popular themes by sets released
sort = df[['Theme','count']].groupby(['Theme'], as_index=False).sum()
top12 = sort.sort_values(['count'], ascending = False).head(12)
top12
```

```python
# make simple bar chart with some variations
plt.bar(top12['Theme'], top12['count'], edgecolor = 'k')
plt.xlabel('Themes', font = 'serif', fontsize = 12)
plt.ylabel('Number of Sets Released', font = 'serif', fontsize = 12)
plt.title('Top 12 Lego Themes by Sets Released', font = 'serif', fontsize = 16)
plt.xticks(rotation = 60, ha = 'right', font = 'serif')
plt.yticks(font = 'serif')

plt.show()

```

## Notice how some of the themes given are not Lego sets. We want to change this to accurately show the top 12 sets

```python
#show the top 20 sets instead
top20 = sort.sort_values(['count'], ascending = False).head(20)
top20
```

```python
#drop themes we don't want
newtop = top20.drop([39, 168, 132, 324, 292, 314, 368, 258],
                    axis = 0)
newtop
```

```python
# same variables, but using Seaborn library. I wanted to include data values inside the bars so I added a function to do this.
ax = sns.barplot(x = 'count', y = 'Theme', data = newtop)
ax.set_xlabel('Total Number of Sets Released', fontsize = 15)
ax.set_ylabel('Theme', fontsize = 15)
ax.set_title('Top 12 Lego Themes by # of Sets', fontsize = 20)

def show_values(axs):
    for p in ax.patches:
        x = p.get_x() + p.get_width() * .98
        y = p.get_y() + p.get_height() * .75
        value = p.get_width().astype(int)
        ax.text(x, y, value, ha= 'right')
        
show_values(0)
```
